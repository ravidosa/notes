### 6.8

```python
t = np.array([0, 26, 50, 76, 100, 126, 150, 176, 200, 226, 250, 276, 300, 326, 350, 376, 400])
V = np.array([3.04, 1.36, 0.28, -0.64, -1.24, -1.76, -2.04, -2.28,  -2.48, -2.60, -2.72, -2.80, -2.84, -2.88, -2.92, -2.96, -3.00])
tt = np.linspace(0, 400, 1000)
plt.plot(tt, 6*np.exp(-tt/79.4794), "c-", label="Prediction")
plt.plot(t, V+3, "bo", label="Data")
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")
plt.title("Voltage vs. Time for $10 nF$ capacitor in RC circuit")
plt.legend()
plt.yscale("log")
plt.show()
#voltage values have been shifted upwards by 3 V so that comparison can be made to the
#expected exponential curve, which is always positive and to work with the log scale, which
#requires positive values only. shifting to a logarithmic scale on the y-axis makes the plot linear
#which is expected as the logarithm "cancels" the exponential
```
