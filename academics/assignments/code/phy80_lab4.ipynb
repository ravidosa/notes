```python
Vth_exp = 3.641 #V
Isc_exp = 2.812 #mA
Rth_exp = 1.294 #k\omega
#expected thevenin open-circuit voltage, short-circuit current, and thevenin resistance
print("The expected Thevenin voltage is " + str(Vth_exp) + " volts.")
print("The expected Thevenin resistance is " + str(Rth_exp) + " kilo-ohms.")
print("The expected short-circuit current is " + str(Isc_exp) + " milliamperes.")

Vth_ms = 3.64 #V
Isc_ms = 2.77 #mA
#measured thevenin open-circuit voltage and short-circuit current
print("The measured Thevenin voltage is " + str(Vth_ms) + " volts.")
print("The measured short-circuit current is " + str(Isc_ms) + " milliamperes.")

R = np.array([0, 470, 1200, 4700])
V = np.array([0, 0.96, 1.74, 2.81, Vth_ms])
I = np.array([Isc_ms, 2.04, 1.43, 0.60, 0])
#measured voltage across and current through load resistors
print("Resistance of varying load resistor:")
print(R)
print("Measured voltages across varying load resistor:")
print(V)
print("Measured current through varying load resistor:")
print(I)

x = np.linspace(0,3.5,100)
plt.plot(x, -Isc_exp/Vth_exp*x + Isc_exp, "k-", label="TE circuit")
plt.plot(x, 1/470*1000*x, "r--", label="470 $\Omega$")
plt.plot(x, 1/1200*1000*x, "g--", label="1.2 $k\Omega$")
plt.plot(x, 1/4700*1000*x, "b--", label="4.7 $k\Omega$")
plt.plot(V, I, "o", label="data")
plt.xlabel("V (volts)")
plt.ylabel("I (mA)")
plt.title("Thevenin Equivalent Circuits")
plt.legend(loc="upper right")
plt.ylim(0, 4)
plt.show()
```
