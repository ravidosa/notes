### 1.1

```python
UPPER = 10
LOWER = 0
STEP = 0.25
tau = 2*np.pi
#initialize range of x values and step size

x = np.arange(LOWER, UPPER, STEP)
y = np.sin(tau * x / 5)
#calculate x and y values for plotting

print("Printing first five entries:\nx\ty\n")
for i in range(5):
    print(str(round(x[i], 2)) + "\t" + str(round(y[i], 2)) + "\n")
    #print first five x and y values, rounded to two decimal places

plt.plot(x, y, "bs", label="sin discrete")
#plot discrete graph with blue (b) squares (s), labeled "sin dicrete"

STEP = 0.01
#smaller step yields smoother graph

x = np.arange(LOWER, UPPER, STEP)
y = np.sin(tau * x / 5)
#recalculate x and y values with new step

plt.plot(x, y, "r--", label="sin smooth")
#plot discrete graph with red (r) dashed lines (--), labeled "sin smooth"

plt.xlabel("x")
plt.ylabel("y")
plt.title(r'$y = \sin(2\pi x/5)$')
plt.legend()
plt.show()
#display graph with x-label "x", y-label "y", title "y = sin(2pix/5)"
```

### 1.2

```python
UPPER = 5
LOWER = -5
STEP = 0.5
tau = 2*np.pi
#initialize range of x values and step size

x = np.arange(LOWER, UPPER, STEP)
y = np.sinc(tau * x / 5)
#calculate x and y values for plotting

print("Printing first five entries:\nx\ty\n")
for i in range(5):
    print(str(round(x[i], 2)) + "\t" + str(round(y[i], 2)) + "\n")
    #print first five x and y values, rounded to two decimal places

plt.plot(x, y, "bo", label="sinc discrete")
#plot discrete graph with blue (b) circles (o), labeled "sinc dicrete"

STEP = 0.01
#smaller step yields smoother graph

x = np.arange(LOWER, UPPER, STEP)
y = np.sinc(tau * x / 5)
#recalculate x and y values with new step

plt.plot(x, y, "r-", label="sinc smooth")
#plot discrete graph with red (r) solid lines (-), labeled "sinc smooth"

plt.xlabel("x")
plt.ylabel("y")
plt.title(r'$y = \frac{\sin(2\pi x/5)}{2\pi x/5}$')
plt.legend()
plt.show()
#display graph with x-label "x", y-label "y", title "y = sin(2pix/5)/(2pix/5)"
```

### 1.3

```python
data = np.loadtxt("../misc/phy80_lab1_hist_data.txt")
#load data from "hist data.txt"

np.shape(data)
#check shape of data (311,)

print("Printing first five entries:\nx\n")
for i in range(5):
    print(str(data[i]) + "\n")
    #print first five data points

NBIN1 = 5
NBIN2 = 20
NBIN3 = 40
NBIN4 = 80
#initialize bin sizes

hx1, edges1 = np.histogram(data, NBIN1)
cbins1 = (edges1[1:] + edges1[:-1])/2
hx2, edges2 = np.histogram(data, NBIN2)
cbins2 = (edges2[1:] + edges2[:-1])/2
hx3, edges3 = np.histogram(data, NBIN3)
cbins3 = (edges3[1:] + edges3[:-1])/2
hx4, edges4 = np.histogram(data, NBIN4)
cbins4 = (edges4[1:] + edges4[:-1])/2
#calculate histograms and bin centers from data for each bin size

fig, axes = plt.subplots(1, 4, figsize=(14, 4))
axes[0].plot(cbins1, hx1, "r-", drawstyle="steps-pre", label=NBIN1)
axes[1].plot(cbins2, hx2, "r-", drawstyle="steps-pre", label=NBIN2)
axes[2].plot(cbins3, hx3, "r-", drawstyle="steps-pre", label=NBIN3)
axes[3].plot(cbins4, hx4, "r-", drawstyle="steps-pre", label=NBIN4)
#plot four horizontally stacked subplots with red (r) solid lines (-)

for i in range(4):
    axes[i].set_xlabel("bins")
    axes[i].set_ylabel("counts")
    axes[i].legend()
    #set x-label as "bins" and y-label as "counts" for each figure
```

```python
NTOT1 = 100
NTOT2 = 1000
BINS = 10
#initialize number of events and bin size

xarr1 = np.random.uniform(size=NTOT1)
xarr2 = np.random.uniform(size=NTOT2)
#generate array of random values between 0 and 1 with 100/1000 values

hx1, edges1 = np.histogram(xarr1, BINS)
cbins1 = (edges1[1:] + edges1[:-1])/2
hx2, edges2 = np.histogram(xarr2, BINS)
cbins2 = (edges2[1:] + edges2[:-1])/2
#calculate histograms and bin centers from data for each bin size

fig = plt.figure()
gs = fig.add_gridspec(2, hspace=0.1)
axes = gs.subplots(sharex=True)
fig.suptitle("Uniform Distribution")
axes[0].errorbar(cbins1, hx1, yerr=np.sqrt(hx1), fmt="ko", label="generated 100 events")
axes[1].errorbar(cbins2, hx2, yerr=np.sqrt(hx2), fmt="ko", label="generated 1000 events")
#plot graphs on same x-axis, with error bars and black (k) circles (o)

axes[0].set_ylim(0, NTOT1 * 0.25)
axes[1].set_ylim(0, NTOT2 * 0.25)
#set limits of y-axis to approximately 1/4 of total number of events

xp = np.linspace(0, 1, 1000)
yp1 = np.full(xp.shape, NTOT1 / BINS)
yp2 = np.full(xp.shape, NTOT2 / BINS)
#prediction for probabilities is number of events divided by bin size

axes[0].plot(xp,  yp1, "r--", label="scaled prediction")
axes[1].plot(xp,  yp2, "r--", label="scaled prediction")
#plot scaled predictions with red (r) dashed lines (--)

axes[0].legend(loc=1, frameon=False)
axes[1].legend(loc=1, frameon=False)

plt.xlabel("x")
axes[0].set_ylabel("entries")
axes[1].set_ylabel("entries")
plt.show()
#display graph with x-label "x", y-label "entries", title "Uniform Distribution"
```
