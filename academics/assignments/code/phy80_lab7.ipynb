### 7.3

```python
#ratio of frequency to cutoff frequency
ff_0 = np.array([1/100, 1/30, 1/10, 1/3, 1, 3, 10, 30, 100])
#v_in for low pass (V)
v_in_l = np.array([1.46, 1.43, 1.43, 1.41, 1.40, 1.37, 1.38, 1.41, 1.44])
#v_out for low pass (V)
v_out_l = np.array([1.46, 1.43, 1.43, 1.34, 1.01, 0.455, 0.141, 0.0489, 0.0167])
#phase for low pass (degrees)
ph_l = np.array([-2.09, -2.63, -6.30, -18.5, -45.3, -70.6, -82.1, -86.4, -91.4])
#v_in for high pass (V)
v_in_h = np.array([1.42, 1.45, 1.41, 1.44, 1.43, 1.37, 1.42, 1.39, 1.43])
#v_out for high pass (V)
v_out_h = np.array([0.0142, 0.0475, 0.140, 0.457, 0.989, 1.30, 1.41, 1.38, 1.43])
#phase for high pass (degrees)
ph_h = np.array([102, 87.7, 85.0, 70.9, 46.2, 19.1, 4.91, 2.10, 1.58])
ffp = np.linspace(0.01, 100, 10000)

plt.plot(ffp, 1/np.sqrt(1+ffp ** 2), "m--", label="Prediction (low-pass)")
plt.plot(ffp, 1/np.sqrt(1+(1/ffp) ** 2), "c--", label="Prediction (high-pass)")
plt.plot(ff_0, v_out_l/v_in_l, "ro", label="Data (low-pass)")
plt.plot(ff_0, v_out_h/v_in_h, "bo", label="Data (high-pass)")
plt.xlabel('$f/f_0$')
plt.ylabel("G(f)")
plt.xscale("log")
plt.title("Gain of RC low-pass and high-pass filters")
plt.legend()
plt.show()
#the prediction matches the data fairly well
```

### 7.4

```python
plt.plot(ffp, -np.arctan(ffp), "m--", label="Prediction (low-pass)")
plt.plot(ffp, np.arctan(1/ffp), "c--", label="Prediction (high-pass)")
plt.plot(ff_0, ph_l*np.pi/180, "ro", label="Data (low-pass)")
plt.plot(ff_0, ph_h*np.pi/180, "bo", label="Data (high-pass)")
plt.xlabel('$f/f_0$')
plt.ylabel('$\phi$ (rad)')
plt.xscale("log")
plt.title("Phase Shift of RC low-pass and high-pass filters")
plt.legend()
plt.show()
#the prediction matches the data fairly well, although it is off a bit at the extreme values
#this makes sense as the oscilloscope did not have a stable measured value for phase
#and these values were measured manually
```

```python
plt.plot(ffp, 20*np.log(1/np.sqrt(1+ffp ** 2)), "m--", label="Prediction (low-pass)")
plt.plot(ffp, 20*np.log(1/np.sqrt(1+(1/ffp) ** 2)), "c--", label="Prediction (high-pass)")
plt.plot(ff_0, 20*np.log(v_out_l/v_in_l), "ro", label="Data (low-pass)")
plt.plot(ff_0, 20*np.log(v_out_h/v_in_h), "bo", label="Data (high-pass)")
plt.xlabel('$f/f_0$')
plt.ylabel("G(f) (dB)")
plt.xscale("log")
plt.title("Gain (in decibels) of RC low-pass and high-pass filters")
plt.legend()
plt.show()
#the prediction matches the data fairly well
```
