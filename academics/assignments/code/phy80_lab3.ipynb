### 3.5

```python
###uncertainties
dV = np.array([0.072, 0.1195, 0.1675, 0.217, 0.2675, 0.3145, 0.364, 0.414, 0.4625, 0.4145])
dI = np.array([0.0256, 0.04, 0.0544, 0.06925, 0.0847, 0.0985, 0.11335, 0.1285, 0.1429,0.15865])
print("Uncertainty in measured voltages:")
print(dV)
print("Uncertainty in measured currents:")
print(dI)
###fractional uncertainties
frV = np.divide(dV, V)
frI = np.divide(dI, I)
print("Fractional uncertainty in measured voltages:")
print(frV)
print("Fractional uncertainty in measured currents:")
print(frI)
#the voltage measurements have a higher fractional uncertainty
```

### 3.6

```python
R1 = 1
plt.plot(V, I, "bo", label="measured")
Vexp = np.linspace(0, 10, 100)
Iexp = Vexp / R1
plt.plot(Vexp, Vexp / R1, "r-", label="expected")
plt.xlabel("voltage across $R_1$ (V)")
plt.ylabel("current through $R_1$ (mA)")
plt.title("Current vs. Voltage for $1k\Omega$ resistor")
plt.legend()
plt.show()
```
