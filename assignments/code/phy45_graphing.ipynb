```python
import matplotlib.pyplot as plt
import numpy as np
import csv

t_arr = np.array([0])
r_arr = np.array([1])
th_arr = np.array([0])

r = 1
th = 0
t = 0

dr = 0
dth = 0
dt = 0.00001
ddr = 0
ddth = 0

g = 9.8
mM = 0.1

while (t < 1):
    ddr = (mM * r * dth ** 2 - g + mM * g * np.sin(th))/(1 + mM)
    ddth = (g * np.cos(th) - 2 * dr * dth)/r
    dr = dr + dt * ddr
    dth = dth + dt * ddth
    r = r + dt * dr
    th = (th + dt * dth) % (2*np.pi)
    t = t + dt
    r_arr = np.append(r_arr, r)
    th_arr = np.append(th_arr, th)
    t_arr = np.append(t_arr, t)

fig, axes = plt.subplots(1, 2, figsize=(14,8))
fig.tight_layout(pad=5.0)
axes[0].plot(t_arr, r_arr, color="red")
axes[1].plot(t_arr, th_arr, color="blue")
axes[0].set_xlabel('Time [s]')
axes[1].set_xlabel('Time [s]')
axes[0].set_ylabel('r [m]')
axes[1].set_ylabel(' [rad]')
axes[0].set_title('r vs. Time')
axes[1].set_title(' vs. Time')
plt.show()
print(min(r_arr))
```

```python
x = 10
y = 1
vx = -3
vy = 0
a = 1
m = 1
dt = 0.01
N = 1000

posx = np.array([])
posy = np.array([])
for i in range(N):
    x = x+vx*dt
    y = y+vy*dt
    Fx = -a/(x**2+y**2)**(3/2)*x
    Fy = -a/(x**2+y**2)**(3/2)*y
    vx = vx+Fx/m*dt
    vy = vy+Fy/m*dt
    t=t+dt
    posx = np.append(posx, x)
    posy = np.append(posy, y)

plt.plot(posx, posy, 'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Particle experiencing Coulomb force (y vs. x)')
plt.legend()
plt.show()
```

```python
tmp = np.arctan2(posy, posx)
tmp[tmp<0] += 2*np.pi
plt.plot(tmp, np.sqrt(posx**2 + posy**2), 'r-')
plt.xlabel('theta')
plt.ylabel('r')
plt.title('Particle experiencing Coulomb force (r vs. theta)')
plt.legend()
plt.show()
```
