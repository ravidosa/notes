```python
#5.1
def powers(a, n):
    #array of numbers from 0 to n
    b = np.arange(0, n + 1, 1)
    #raise a to power of array
    return a ** b
print(powers(2, 4))
```

```python
#5.2
a = np.array([])
print(a)
a = np.append(a, 1)
print(a)
a = np.append(a, 3)
print(a)
a = np.append(a, 4)
print(a)
#initializes array, then appends 1, 3, and 4
```

```python
#5.3
def recur(n, x):
    b = np.array([x])
    for i in range(1, n):
        x = 2 * x + 1
        b = np.append(b, x)
        #calculates next term and appends n - 1 times
    return b
print(recur(5, 1))
```

```python
#5.4
UPPER = 5
LOWER = -5
tau = 2*np.pi
x = np.linspace(LOWER, UPPER, 1000)
sc = np.sinc(x)
plt.plot(x, sc, "g-")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sinc(x)")
plt.show()
#graphs sinc(x) = sin(x)/x between -5 and 5
```

```python
#5.5
def logmap(p, r):
    return r * p * (1 - p)
p = 0.1
print(p)
for i in range(5):
    p = logmap(p, 3)
    #calculates next term
    print(p)
```

```python
#5.6
p = 0.1
b = np.array([p])
for i in range(6):
    p = logmap(p, 3)
    b = np.append(b, p)
    #calculates next term and appends
print(b)
```

```python
#5.7
x = np.arange(0, 40, 1)
p = 0.1
b = np.array([p])
for i in range(39):
    p = logmap(p, 2.8)
    b = np.append(b, p)
    #calculates next term
#print(b)
plt.scatter(x, b)
plt.xlabel("x")
plt.ylabel("p_x")
plt.title("Logistics Map for p0 = 0.1, r = 2.8")
plt.show()
#graphs time evolution of logistics map for p0 = 0.1 and r = 2.8
```
