### 8.6

```python
#lissajous fish pattern using scientific python
t = np.linspace(0,1,1000)
w = 2*np.pi
x = 3*np.sin(2*w*t)
y = 3*np.sin(3*w*t+np.pi/4*5)
plt.plot(x,y)
plt.axis("equal")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lissajous Fish Pattern")
plt.show()
```

### 8.7

```python
#lissajous parabola pattern using scientific python
t = np.linspace(0,1,1000)
w = 2*np.pi
x = 3*np.sin(1*w*t+np.pi/4)
y = 3*np.sin(2*w*t)
plt.plot(x,y)
plt.axis("equal")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lissajous Parabola Pattern")
plt.show()
```

### 8.8

```python
#lissajous crown pattern using scientific python
t = np.linspace(0,1,1000)
w = 2*np.pi
x = 3*np.sin(1*w*t)
y = 1*np.sin(4*w*t)
plt.plot(x,y)
plt.axis("equal")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lissajous Crown Pattern")
plt.show()
```
